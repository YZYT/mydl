```python
%matplotlib inline
import torch
import torchvision
import torch.nn as nn
import matplotlib
from torch.utils import data
import matplotlib.pyplot as plt
from torchvision import transforms
from d2l import torch as d2l
import numpy as np
from IPython import display

d2l.use_svg_display()
d2l.init_Seed(233)
device = d2l.get_device()
```

```python
def softmax(X):
    X_exp = torch.exp(X)
    partition = X_exp.sum(dim=1, keepdim=True)
    return X_exp / partition

def cross_entropy(y_hat, y):
    return -torch.log(y_hat[range(len(y)), y])
```

```python
config = {
    "batch_size": 100,
    "n_epochs": 50,
    "optimizer": "SGD",
#     "optimizer": "Adam",
    "optim_hparas": {
        'lr': 0.03,
        'momentum': 0.9,
        'nesterov':True
    },
    
    'lr_scheduler': 'StepLR',
    'lr_sched_hparas':{
        'step_size': 1,
        'gamma': 0.95
    }
}
```

```python
class SoftmaxReg(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.net = nn.Sequential(nn.Flatten(), nn.Linear(784, 256), nn.ReLU(), nn.Linear(256, 128), nn.ReLU(), nn.Linear(128, 10))
        
        def init_weights(m):
            if type(m) == nn.Linear:
                nn.init.normal_(m.weight, std=0.001)

        self.net.apply(init_weights)

    def forward(self, X):
        return self.net(X)
```

```python
tr_set, te_set = d2l.load_data_fashion_mnist(batch_size=config["batch_size"])
```

```python
model = SoftmaxReg().to(device)
loss = nn.CrossEntropyLoss()
optimizer = getattr(torch.optim, config["optimizer"])(
    model.parameters(), **config["optim_hparas"]
)
lr_scheduler = getattr(torch.optim.lr_scheduler, config["lr_scheduler"])(optimizer, **config["lr_sched_hparas"])

min_loss, max_acc = d2l.train_clf(model, tr_set, te_set, loss, optimizer, lr_scheduler, device, config['n_epochs'])

print(min_loss, max_acc)
```

```python
model.to('cpu')
for X, y in te_set:
    trues = d2l.get_fashion_mnist_labels(y)
    preds = d2l.get_fashion_mnist_labels(model(X).argmax(axis=1))
    titles = [true + '\n' + pred for true, pred in zip(trues, preds)]
    d2l.show_images(X[:18].reshape(18, 28, 28), 2, 9, titles)
    break
```
